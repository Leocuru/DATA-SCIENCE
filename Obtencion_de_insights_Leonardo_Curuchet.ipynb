<center><h2>ANALISIS DE LAS VENTAS DE BEBIDAS</h2></center>



El sector de bebidas es una industria diversificada y en constante evolución que atiende a una amplia gama de consumidores y preferencias. Esta investigación se centra en un análisis de ventas de una empresa que se especializa en la distribución de bebidas, tanto alcohólicas como no alcohólicas, a lo largo de los años 2020, 2021, 2022 y 2023. El estudio abarca una variedad de variables clave, incluyendo tipos de bebidas, marcas, vendedores responsables, supervisores encargados, categorías de bebidas, calibres de productos y tipos de clientes atendidos.

El análisis exploratorio de datos revela patrones notables en las ventas a lo largo de las estaciones del año y destaca la creciente demanda durante la temporada de verano en comparación con otras estaciones. Además, se observa que las bebidas sin alcohol muestran una estabilidad a lo largo del año en contraste con las bebidas alcohólicas.

Este estudio también investiga la preferencia de calibres de productos por parte de los clientes y cómo varía según los diferentes subcanales de marketing a los que pertenecen. Además, se analiza la influencia de los vendedores en las ventas.

Los resultados de este análisis proporcionan información valiosa para la toma de decisiones empresariales y estrategias de marketing. A medida que la industria de bebidas continúa evolucionando, comprender estos patrones y preferencias es esencial para satisfacer las necesidades cambiantes de los consumidores y mantener el éxito en el mercado.

```python
def asignar_estacion(Mes):
    if Mes in ['Diciembre', 'Enero', 'Febrero']:
        return 'Verano'
    elif Mes in ['Marzo', 'Abril', 'Mayo']:
        return 'Otoño'
    elif Mes in ['Junio', 'Julio', 'Agosto']:
        return 'Invierno'
    else:
        return 'Primavera'

df['estacion'] = df['Mes'].apply(lambda x: asignar_estacion(x))
df.head()
```

```python
# Calcula la suma de bultos vendidos por cada vendedor
ventas_por_vendedor = df.groupby('Sector de Venta')['Cantidad Total en Bultos'].sum()

# Calcula los porcentajes de ventas por calibre para cada vendedor
tabla_porcentajes_calibre = df.pivot_table(index='Sector de Venta', columns='Calibre', values='Cantidad Total en Bultos', aggfunc='sum', fill_value=0)
tabla_porcentajes_calibre = tabla_porcentajes_calibre.div(tabla_porcentajes_calibre.sum(axis=1), axis=0) * 100

# Combina las dos tablas en una sola
tabla_final = ventas_por_vendedor.to_frame().join(tabla_porcentajes_calibre)

# Renombra la columna de bultos vendidos
tabla_final.rename(columns={'Cantidad Total en Bultos': 'Bultos Vendidos'}, inplace=True)

# Ordena las columnas (calibres) en función del volumen de ventas
tabla_final = tabla_final[tabla_final.sum().sort_values(ascending=False).index]

# Ordena las filas (vendedores) en función de las ventas totales
tabla_final = tabla_final.sort_values(by='Bultos Vendidos', ascending=False)

# Aplica el formato a la tabla
tabla_final = tabla_final.applymap(lambda x: f'{x:.1f}%' if '%' in str(x) else f'{x:.1f}')

# Imprime la tabla final
print(tabla_final)
```

HIPOTESIS 3: "El comportamiento de las marcas mas importantes es mas estable durante el año que las menos importantes".

DETERMINAMOS QUE MARCAS SON IMPORTANTES Y CUALES NO, CUANDO LA PARTICIPACION EN LAS VENTAS SUPERAN EL 20% DEL TOTAL

```python
# 1. Calcular la cantidad total en bultos por marca
total_bultos_por_marca = df.groupby('Marca')['Cantidad Total en Bultos'].sum().reset_index()

# 2. Identificar las marcas importantes (superan el 20%)
umbral_importancia = 0.20  # 20%
marcas_importantes = total_bultos_por_marca[total_bultos_por_marca['Cantidad Total en Bultos'] > (umbral_importancia * total_bultos_por_marca['Cantidad Total en Bultos'].sum())]

# 3. Crear una nueva columna de importancia en el DataFrame original
df['Marca Importante'] = df['Marca'].isin(marcas_importantes['Marca']).astype(int)
```

```python
import matplotlib.pyplot as plt

# Filtrar los datos para marcas importantes y no importantes
marcas_importantes = df[df['Marca Importante'] == 1]
marcas_no_importantes = df[df['Marca Importante'] == 0]

# Agrupar y sumar las cantidades mensuales para marcas importantes
datos_importantes = marcas_importantes.groupby('Mes')['Cantidad Total en Bultos'].sum()

# Agrupar y sumar las cantidades mensuales para marcas no importantes
datos_no_importantes = marcas_no_importantes.groupby('Mes')['Cantidad Total en Bultos'].sum()

# Crear un gráfico de líneas para marcas importantes
plt.figure(figsize=(12, 6))
plt.plot(datos_importantes.index, datos_importantes.values, label='Marcas Importantes')
plt.title('Acumulado Mensual de Marcas Importantes')
plt.xlabel('Mes')
plt.ylabel('Acumulado de Cantidad Total en Bultos')
plt.legend()
plt.grid()
plt.show()

# Crear un gráfico de líneas para marcas no importantes
plt.figure(figsize=(12, 6))
plt.plot(datos_no_importantes.index, datos_no_importantes.values, label='Marcas No Importantes')
plt.title('Acumulado Mensual de Marcas No Importantes')
plt.xlabel('Mes')
plt.ylabel('Acumulado de Cantidad Total en Bultos')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Crear el modelo
model = LinearRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Predecir las ventas en el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular métricas de evaluación
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f'R²: {r2}')
print(f'MSE: {mse}')
print(f'MAE: {mae}')
```

REGRESION POLINOMICA

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Crear un modelo de regresión polinómica de grado 2
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())

# Entrenar el modelo polinómico
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Hacer predicciones con el modelo polinómico
y_pred = model.predict(X_test)

# Calcular R²
r2 = r2_score(y_test, y_pred)

# Calcular MSE
mse = mean_squared_error(y_test, y_pred)

# Calcular MAE
mae = mean_absolute_error(y_test, y_pred)

print(f'R²: {r2}')
print(f'MSE: {mse}')
print(f'MAE: {mae}')
```

EL R2 ES MEJOR QUE EN LA REGRESION LINEAL, PERO AUN ESTA LEJOS DE SER UN MODELO QUE DE CONFIANZA, VAMOS POR OTRO MODELO MAS
